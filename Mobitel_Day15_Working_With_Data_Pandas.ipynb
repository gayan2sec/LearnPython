How to work with Data

Structured Data: 

Data that follows a specific and predefined format or schema. It is typically organized into well-defined rows and columns, resembling a table-like structure. Structured data is commonly found in relational databases and can be easily queried and analyzed using SQL (Structured Query Language).

Unstructured Data: 

Data that does not adhere to a fixed schema and lacks a predefined structure. Unstructured data can be more complex and includes information like text documents, images, audio recordings, video files, social media posts, and other free-form content.

Semi-structured 

Semi-structured data refers to data that does not have a fixed schema or predefined structure like traditional relational databases. Unlike structured data, which conforms to a rigid schema with well-defined rows and columns, semi-structured data allows for more flexibility in representing and storing information.

Data Nesting: It allows for nesting and hierarchies, which means data can be organized in a tree-like or hierarchical structure, typically using formats like JSON (JavaScript Object Notation), XML (eXtensible Markup Language), or YAML (YAML Ain't Markup Language).
